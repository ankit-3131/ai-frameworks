```python
# !pip install -U "langchain[google-genai]"
import os
import getpass
from pydantic import BaseModel, Field
from typing import Literal # Corrected import for Literal
from langchain.chat_models import init_chat_model
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google AI API key: ")
model_without_strutured = init_chat_model("google_genai:gemini-3-flash-preview")


class Poem(BaseModel):
  verse: str = Field(description="poem verse")
  tone: Literal["happy","sad","melancholy"] = Field(description = "if any other specify that")

model = model_without_strutured.with_structured_output(Poem, include_raw = True)

result = model.invoke("write a poem about watching moon at the shore")
print(result)
```
