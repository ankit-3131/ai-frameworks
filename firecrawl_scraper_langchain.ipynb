```python
os.environ["FIRECRAWL_API"] = userdata.get('FIRECRAWL_API')
loader = FireCrawlLoader(
    api_key=userdata.get('FIRECRAWL_API'), url="https://leetcode.com/u/ankit_3131/", mode="scrape"
)

pages = []
for doc in loader.lazy_load():
    pages.append(doc)
print(pages)
```
